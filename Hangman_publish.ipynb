### Game code

```python
chosen_word = ''
user_name = ''
user_topic_choice = ''

#Phase 1: user telling us his/her/etc name.
user_name = str(input('What is your name? '))

#we are introducing game and category option.
print(f'''
Hello {user_name}, welcome to Hangman 2000: The Ultimate Letter Elimination Experience!

You can choose between 4 topics to play: countries, foods, animals, films
in our next update you can choose kardashians, where you could guess your favorite kardashian name!

Rules are simple, you have 5 chance to make mistake, after that Game over!
==========
''')
#in this loop we are letting them choose the topic, if they choose wrongly, they have to choose again, when they choose, the program assigns a random word from that chosen topic and lets users know how many letters are in the word and a display of it.
while True:
  user_topic_choice = str(input('please choose your topic: \n').lower())
  if user_topic_choice in topic_list:
    topic_choice = topic_list[user_topic_choice]
    chosen_word = rd.choice(topic_choice)
    chosen_word = chosen_word.lower()
    char_count = sum(1 for char in chosen_word if char != " ") #we want to exclude spaces between words to get accurate letter count.
    display_word = ''
    for char in chosen_word: #we want to show letters with '-' and spaces with ' '.
      if char == ' ':
        display_word += " "
      else:\
        display_word += "-"
    print(f'\nGood Choice,{user_name}!' )
    print(f'You have to guess {char_count} letter word related to {user_topic_choice}\n')
    print(display_word)
    print('\ngood luck!\n')
    print('==========')
    break
    #after they chose right category, the loop will break, otherwise it will ask them again and again.
  else:
    print(f'Sorry {user_name}, you can only choose between countries, foods, animals or films yet.')

#here game starts and we need to reset this variables so each time player is starting new game, they wouldnt store previous values.
chosen_letter = str(input('Guess your first letter! ').lower())
wrong_count = 0
display_word = ''
chosen_letters = []#for displaying right guessed letters along with dash and spaces, we need a list to store right guessed letters.
wrong_letters = []#to show player wrong letters guessed, we are storing them in a list.



while True:
  display_word = ''
  if chosen_letter in chosen_word: #validating the letter player chose
    chosen_letters.append(chosen_letter) #adding to list of letters player chose
    for char in chosen_word: #this loops helps us to show the word with letters player have guessed and remaining ones with '-'
      if char in chosen_letters:
        display_word += char
      elif char == ' ':
        display_word += ' '
      else:
        display_word += '-'
    print('\nCongratulation! you guessed a letter right!\n')
    print(display_word)
    print('==========')
    if chosen_word.lower() == display_word.lower(): #validating if they won the game
      print('='*10)
      print(f'Kudos for you {user_name}, you won!')
      print('='*10)
      break
    else:
      chosen_letter = str(input('Guess your next letter! ').lower()) #asking again and again for letter until they loose or won.
  else: #wrong guess section
    wrong_count += 1
    wrong_letters.append(chosen_letter)
    if wrong_count < 6: #here we are warning them and showing wrong guessed letters.
      print('\noops, better luck next time\n')
      print(hangman_draw[wrong_count])
      print('Think about this man\'s family!')
      print(f'You have geussed {wrong_count} times wrongly\n')
      print(f'Wrong letters: {wrong_letters}\n')
      print('==========')
      chosen_letter = str(input('Guess your next letter! ').lower())
    if wrong_count ==  6: #here they loose.
      print(hangman_draw[wrong_count])
      print(f'Sorry {user_name}, because of you and cruel world of letters, there is an empty chair in dinner table somewhere')
      print('Maybe with kardashians topic you will do better.')
      print(f'You had to guess {chosen_word}')
      break

```
